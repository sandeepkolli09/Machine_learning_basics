#                                                          MACHINE LEARNING

Machine Learning (ML) is a branch of artificial intelligence that enables systems to learn and improve from experience without being explicitly programmed. It involves using algorithms to analyze data, identify patterns, and make predictions or decisions.

## supervised learning:

### what is supervised learning?
Supervised learning is a type of machine learning where a model is trained on labeled data. In this process, the algorithm learns to map input features (independent variables) to the correct output (dependent variable) by using example input-output pairs provided during training.

#### Types:
Classification: Predicting a discrete label (e.g., "spam" or "not spam").

Regression: Predicting a continuous value (e.g., house prices).


### Regression:
Regression is a statistical and machine learning technique used to model and analyze the relationships between a dependent variable (often called the target or output) and one or more independent variables (also called features or inputs). The main goal of regression is to predict a continuous output value based on the input features.

#### 1)Linear regression:


Linear Regression is a  statistical technique which gives the  relationship between a dependent variable (output) and one or more independent variables (inputs) by fitting a straight line through the data points. 

#### Types of Linear Regression:
1.Simple linear regression
2.Multiple linear regression

Formula :
     
     "y=mx+c"-- is the equation of the regression line, which is the best-fit line that minimizes the distance between the actual data points and the predicted values y.
     
 where y--->dependent variable or the predicted outcome
  
       m--->m is the slope of the line, which indicates how much y changes for a one-unit increase in x.
       
       x--->input value or independent value
       
       c--->c is the y-intercept, the point where the line crosses the y-axis.

Simple Linear regression with code:

```python
import pandas as pd
from  sklearn import linear_model
import numpy as np
import matplotlib.pyplot as plt
```

```python
file=pd.read_csv("files/homeprices.csv")  #data set reading
```

```python
file.head()
```

The dataset working with consists of two columns:

Area (square feet): This is the independent variable (denoted as x in the linear regression equation), which represents the size of a house in square feet.

Price (in dollars): This is the dependent variable (denoted as y in the equation), which represents the price of the house. The goal of the linear regression model is to predict the house price based on its area.

he linear regression equation you are working towards will be of the form:

Price=0+1Area

0 is the intercept, representing the price when the area is zero.
1 is the slope representing how much the house price increases for each additional square foot of area.

Now that we have plotted the dataset, we can visually analyze the relationship between the two variables: Area (in square feet) and Price (in dollars).

```python
plt.scatter(file.area,file.price,color="red",marker="o")
```

#### by observation:
The scatter plot reveals a clear upward trend, indicating a positive correlation between area and price, where larger areas typically correspond to higher prices.

let fit a line to understand the linearity 

```python
slope,intercept=np.polyfit(file['area'],file['price'],1) 
##The np.polyfit function from the NumPy library is used to perform polynomial curve fitting, which in this case is being applied to linear regression (a first-degree polynomial).
```

```python
line=slope*file['area']+intercept  ##y=mx+c
```

```python
plt.scatter(file.area, file.price, color='blue')
plt.plot(file.area, line, color='red', label='Line of Best Fit')
```

```python
model=linear_model.LinearRegression()
##LinearRegression() is a class within the linear_model module that implements linear regression.
##When you call this class, you are creating an instance of the linear regression model
```

```python
new=file.drop('price',axis='columns')
new
price=file.price
price
```

```python
model.fit(new,price)
```

model.fit() is a method that trains the linear regression model using the provided data.
By executing this line, you are instructing the model to analyze the relationship between the input features (areas) and the target variable (prices)

```python
model.predict([[3402]]) ## testing the output
```

```python
model.coef_ ##model.coef_ is an attribute of the fitted linear regression model that contains the coefficients (slopes) of the features used in the model.
```
